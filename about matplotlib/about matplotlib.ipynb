# Font

````python
font_settings = matplotlib.font_manager.FontProperties()
font_settings.set_family("Dejavu Sans")
# 'serif', 'sans-serif', 'cursive', 'fantasy', 'monospace'
font_settings.set_size(20)
font_settings.set_style("normal")
font_settings.set_weight("bold")

fig, ax = plt.subplots(figsize=(5, 3))

data = [10, 24, 30, 50, 40]
ax.plot(data)
ax.set_title("title", fontproperties=font_settings)
plt.show()
````

````python
plt.rcParams["font.size"] = 20

fig, ax = plt.subplots(figsize=(10, 3))

data_x = [0.001, 0.01, 0.1, 1, 10]
data_y = [10, 24, 30, 50, 40]

ax.plot(data_x, data_y)
ax.set_title("한글 그래프")
ax.set_xscale("log")

plt.show()
````

지수에 마이너스는 unicode로 변환 시 U+207B를 가리키나, matplotlib의 지수 마이너스는 U+2212로 U+2212를 지원하지 않는 font는 이전으로 자동변환되지 않아 제대로 나오지 않는다. 쉬운 해결책은 지원하는 font를 이용하는 것이다.

````python
plt.rcParams["font.size"] = 20

fig, ax = plt.subplots(figsize=(10, 3))

ax.plot(data_x, data_y)

plt.xticks(fontname="Liberation Sans")
ax.set_title("한글 그래프")
ax.set_xscale("log")

plt.show()
````
